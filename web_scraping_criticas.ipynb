**Importando as Bibliotecas**

```python
import requests
from bs4 import BeautifulSoup
```

**Esse código abre o site, baixa o código da página em HTML e organiza esse código com o BeautifulSoup e mostra ele de forma mais arrumada na tela.**

```python
#link = 'https://www.omelete.com.br/filmes/criticas/divertida-mente'
#link = 'https://www.adorocinema.com/filmes/filme-196960/criticas-adorocinema/'
link = 'https://www.planocritico.com/critica-divertida-mente/'
requisicao = requests.get(link)
print(requisicao)
#print(requisicao.text)

site = BeautifulSoup(requisicao.text, "html.parser")
print(site.prettify())
```

**Esse comando mostra o título do site.**

```python
print(site.title)
```

**Esse código procura todos os parágrafos (p) da página e, para cada um, mostra apenas o texto limpo, sem as tags HTML, trazendo os comentarios dos Criticos.**

```python
pesquisa = site.find_all("p")
for p in pesquisa:
    print(p.get_text(strip=True))
```
